## Gender-wise Purchasing Power
### Which gender contributes more to total sales during Diwali?

```python
gender_data=new_data.groupby('Gender')['Amount'].sum().reset_index()
gender_data
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=gender_data, x='Gender', y='Amount')
plt.title('Gender-wise Total Purchase Amount During Diwali')
plt.xlabel('Gender')
plt.ylabel('Total Purchase Amount')
plt.show()
```

## Interpretation
#### here it shows that Female contributes twice than male in dewali sales 

## Age Group Spending Pattern
###  Which age group spends the most during the Diwali festival?


```python
age_data=new_data.groupby('Age Group')['Amount'].sum().reset_index()
age_data
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=age_data,x='Age Group',y='Amount')
plt.title('Age-wise Total Purchase Amount During Diwali')
plt.xlabel('Age Group')
plt.ylabel('Total Purchase Amount')
plt.show()
```

## Interpretation
#### here 26-35 age group purchase more where as below 17 purchase less in dewali

## State-wise Sales Performance
### Which states generate the highest total sales?

```python
state_data=new_data.groupby('State')['Amount'].sum().reset_index()
state_data
```

```python
sorted_state_data=state_data.sort_values(by='Amount').reset_index()
sorted_state_data
```

```python
sorted_state_data.drop(columns="index")
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=sorted_state_data,x='Amount',y='State')
plt.title('State-wise Total Purchase Amount During Diwali')
plt.ylabel('State')
plt.xlabel('Total Purchase Amount')
plt.show()
```

## Interpretation
### here it shows that uttar pradesh purchase more in diwali and telangana purchase less

## Product Category Demand
### Which product categories are most popular during Diwali?

```python
demand=new_data.groupby('Product_Category')['Orders'].sum().reset_index()
demand
```

```python
total=new_data['Orders'].sum()
print(total)
```

```python
demand['orders_percentage']=(demand['Orders']/total)*100
demand
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=demand,x='orders_percentage',y='Product_Category')
plt.title('product categories which are most popular during Diwali')
plt.ylabel('Product category')
plt.xlabel('orders_percentage')
plt.show()
```

## Occupation vs Purchase Amount (The “Money Chart”)
### How does customer occupation influence purchase amount?

```python
com_data=new_data.groupby('Occupation')['Amount'].sum().reset_index()
com_data
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=com_data,x='Amount',y='Occupation')
plt.title('customer occupation vs purchase amount')
plt.xlabel('Purchase amount')
plt.ylabel('Occupation')
plt.show()
```
