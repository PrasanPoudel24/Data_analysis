# Merging
- it is based on common key (or column) and simmilar to SQL joins.
- its used when we want to allign data from two differnet datasets based on shared values.(eg: ID,names)
# joiining
- a special case of merging that combines DFs based on their index or key column
- it is typically used when DFs share a common index
# concatenation
- Stacks DFs either vertically (adding rows) or horizontally(adding columns)
- It is used when combing datasets with similar structues without matching keys

# Merge Practical

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips['customer_id']=pd.Series(np.arange(1,245))
tips
```

```python
tips.head()
```

```python
customer_data = pd.DataFrame({
    'customer_id':[1,2,3,4,5],
    'name':['Geeta','Ram','Hari','Shyam','Rupa']
})
customer_data
```

```python
tips.merge(customer_data,on='customer_id',how='inner')
```

```python
find_tips=pd.merge(tips,customer_data,on='customer_id',how='left')
find_tips
```

```python
customer_data_conflict=pd.DataFrame(
    {
    'customer_id':[1,2,3,4,5],
    'name':['Geeta','Ram','Hari','Shyam','Rupa'],
        'size':['small','medium','large','small','medium']
    }
)
customer_data_conflict
```

```python
tips.merge(customer_data_conflict,on='customer_id',how='inner',suffixes=['_tips','_customer'])
```

# merge from multiple columns

```python
shift_data=pd.DataFrame({
    'day': ['Sun','Sun','Fri','Sat'],
    'time':['Dinner','Lunch','Dinner','Lunch'],
    'shift_manager':['Geeta','Ram','Hari','Shyam']
})
shift_data
```

```python
tips.merge(shift_data,on=['day','time'],how='left')
```

# Join

```python
tips_subset=tips.set_index('customer_id')
customer_data_subset= customer_data.set_index('customer_id')
```

```python
tips_subset.join(customer_data_subset,on='customer_id',how='inner')
```

```python
customer_data_conflict_subset=customer_data_conflict.set_index('customer_id')
```

```python
tips_subset.join(customer_data_conflict_subset,lsuffix='_tips',rsuffix='_customers')
```

```python
tips_subset.loc[[1,2,3,6,7]].join(customer_data_subset)
```

```python
tips_subset.loc[[1,2,3,6,7]].join(customer_data_subset,how='outer')
```

```python
tips_subset
```

```python
customer_data
```

```python
# index amd column
tips_subset.merge(customer_data , right_on='customer_id',left_index=True)
```

```python
tips_subset.merge(customer_data_conflict, right_on='customer_id',left_index=True)
```

# Concatenation

```python
tips=sns.load_dataset('tips')
tips
```

# split tips into 3 parts

```python
tips_1 = tips.iloc[:3]
tips_2 = tips.iloc[3:6]
tips_3 = tips.iloc[6:9]
tips_1
```

```python
tips_2
```

```python
# concatenate vertically with keys
vertical_concat=pd.concat([tips_1,tips_2,tips_3],axis=0,keys=['Part1','Part2','Part3'])
vertical_concat
```

```python
vertical_concat.index
```

```python
vertical_concat.loc['Part1']
```

```python
extra_data=pd.DataFrame({
    'customer_id':[1,2,3,4,5],
    'rating':[4,5,2,3,2]
}, index= tips.head().index)
extra_data
```

```python
pd.concat([tips.head(),extra_data],axis=1, join='outer')
```

```python
# Ignore index
pd.concat([tips_1,tips_2,tips_3],axis=0,ignore_index=True)
```
