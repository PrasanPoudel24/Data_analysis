```python
import numpy as np
import pandas as pd
```

# DateTime Operation
- It is very important operation for time series analysis

# create or change type to datetime
- .to_datetime()

```python
data={
    'Date' :['2023-01-01','2023-01-02','2023-01-03','2023-01-04'],
    'Sales' : [40000,35000,50000,55000]
}
df=pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df['date']=pd.to_datetime(df.Date)
```

```python
df
```

```python
# date_range(start,end,freq=)
new_date=pd.date_range('2023-01-01','2024-01-01',freq='ME')
new_date
```

# Extracting Date/Time components
- .dt accessor

```python
print("Year",df.date.dt.year)
```

```python
df['Year']=df.date.dt.year
```

```python
df['month']=df.date.dt.month
```

```python
df['day']=df.date.dt.day
```

```python
df['hour']=df.date.dt.hour
```
