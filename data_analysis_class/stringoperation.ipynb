```python
titanic=pd.read_csv('titanic.csv')
titanic
```

```python
titanic.Name
```

```python
#split text : .split()
titanic.Name.str.split(',')
```

```python
#.get() : to access index value
titanic['Last_Name']=titanic.Name.str.split(',').str.get(0)
titanic
```

```python
titanic.Last_Name.str.len()
```

```python
titanic['First_name']=titanic.Name.str.split(',').str.get(1)
titanic
```

```python
titanic.Name.str.split(',').str.get(1).str.len()
```

```python
# strip(),rstrip(),lstrip()

titanic.Name.str.split(',').str.get(1).str.strip().str.split(' ')
```

```python
titanic[['Title','first_name_no_Title']]=titanic.Name.str.split(',').str.get(1).str.strip().str.split(' ',n=1,expand= True)
```

```python
titanic
```

```python
# replace content
titanic.Title.value_counts()
```

```python
titanic.Title=titanic.Title.str.replace('Ms.','Miss.')
```

```python
titanic.Title.value_counts()
```

```python
titanic.Last_Name[titanic.Last_Name.str.endswith('a')]
```

```python
# find() : return lowest index point
# -1 defines there is no substring
titanic.Last_Name.str.find('a')
```

```python
# slice(start,stop,step) : create substring by slicing string
titanic.first_name.str.slice(2,15,2)
```
