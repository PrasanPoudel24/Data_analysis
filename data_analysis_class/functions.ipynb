### Apply()
- it is used to apply same methods to all element of series at once

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
tips.info()
```

```python
tips.describe()
```

```python
# categories total bill into low ,mid and high
```

```python
def billcategory(amt):
    if amt<10:
        return "LOW"
    elif amt>10 and amt<30:
        return "mid"
    else:
        return "high"
```

```python
tips['bill_category']=tips.total_bill.apply(billcategory)
```

```python
tips['bill_category']=tips.total_bill.apply(billcategory)
tips
```

```python
tips['size'].apply(lambda x: "group" if x>1 else "single")
```

```python
tips.bill_category.value_counts()
```

# Grouping and Aggregation
- Grouping : group data by some columns or parameters.
- aggregation : provides aggregate values like sum,mean,min
- methods:
     - groupby : groupping
     - agg : apply multiple aggregate methods
     - pivot : multi-dimensional view

# total sales by day.

```python
# .groupby()
```

```python
tips.groupby('day',observed=True)[['total_bill']].sum()
```

```python
# /  : use when code becomes long
sales_by_day=tips.groupby('day',observed=True)[['total_bill']].sum().sort_values(by='total_bill').reset_index()
sales_by_day
```

```python
sales_by_day.plot(kind='bar')
```

```python
sales_by_day.plot(kind='hist')
```

```python
tips
```

```python
tips.groupby(['day','time'],observed=True)[['total_bill']].sum()
```

```python
tips.groupby(['day','time'],observed=True)[['total_bill']].mean()
```

```python
tips.groupby(['day','time'],observed=True)[['total_bill']].agg(['sum','mean'])
```

# convert male ,female into men and women

```python

mapping_day={'Male':'men','Female':'women'}
tips['new_sex']=tips.sex.map(mapping_day)
```

# categories tips as gorgeous for morethan 5, standard for less than 5

```python
tips["new_cat"]=tips['tip'].apply(lambda x: "gorgeous" if x>5 else "standard" )
```

# calculate tip percentage by their bill amount

```python
tips["percentage_on_tip"]=tips["tip"]/tips["total_bill"]*100
```

```python
tips
```

```python
tips[['total_bill','tip','percentage_on_tip']].applymap(lambda x: f'{x:.2f}')
```

# pivot_table
- it provides values default in mean value

```python
pd.pivot_table(tips,values=['tip','total_bill'],index=['sex','smoker'],columns=['day'],observed= True,aggfunc={'total_bill':'mean','tip':'mean'}\
              , margins=True , margins_name ="total")
```
