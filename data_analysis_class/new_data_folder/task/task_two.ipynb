```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ipl=pd.read_csv('IPL Matches 2008-2020.csv')
ipl
```

# `Question 1 :` Which team has the highest win percentage across seasons?

```python
ipl.team1.value_counts()
```

```python
ipl.team2.value_counts()
```

```python
team1_counts = ipl.team1.value_counts()
team2_counts = ipl.team2.value_counts()
matches_played = team1_counts.add(team2_counts, fill_value=0)
matches_played
```

```python
plt.figure(figsize=(10,5))
new1=matches_played.sort_values()
new1.plot(kind='barh')
plt.title("Matches Played of IPL Teams (2008-2020)", fontsize=14)
plt.xlabel("Matches Played")
plt.ylabel("Teams")
plt.show()
```

```python
match_won=ipl.winner.value_counts()
match_won
```

```python
win_per=(match_won/matches_played)*100
win_per
```

```python
plt.figure(figsize=(10,6))
new=win_per.sort_values()
new.plot(kind='barh')

plt.title("Win Percentage of IPL Teams (2008-2020)", fontsize=14)
plt.xlabel("Win Percentage (%)")
plt.ylabel("Teams")
plt.show()
```

# `Question 4 :` Does winning the toss significantly increase the chances of winning a match?

```python
ipl.head()
```

```python
total_matches = ipl[ipl['winner'].notna()].shape[0]
total_matches 
```

```python
toss_match_winner=ipl[ipl['toss_winner']==ipl['winner']].shape[0]
toss_match_winner
```

```python
toss_advantage = (toss_match_winner / total_matches) * 100

print(f"Total Matches: {total_matches}")
print(f"Toss Winner also won: {toss_match_winner}")
print(f"Toss Advantage %: {toss_advantage:.2f}")
```

#  `Question 5 :` Which toss decision (bat/field) leads to more victories?


```python
ipl.columns
```

```python
ipl.toss_decision.value_counts()
```

```python
toss_match_wins = ipl[ipl['toss_winner'] == ipl['winner']]
toss_match_wins
```

```python
decision=toss_match_wins['toss_decision'].value_counts()
decision
```

```python
decision.plot(kind='bar', color=["orange", "green"], figsize=(6,4))


plt.title("Toss Decision vs Victories", fontsize=14)
plt.ylabel("Number of Matches Won")
plt.xlabel("Toss Decision")

plt.show()
```
