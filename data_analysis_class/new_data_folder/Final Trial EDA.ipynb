# importing libraries

```python
# Standard imports
import pandas as pd
import numpy as np

# Visualization library imports
import matplotlib.pyplot as plt
import seaborn as sns

# A special command to make plots appear inline in Jupyter Notebook
%matplotlib inline

# Set a default style for our plots for a nice look
sns.set_style('whitegrid')
```

# Import Dataset

```python
trial = pd.read_csv('final_trial.csv')
trial
```

```python
trial.drop(columns='Unnamed: 0', inplace =True)
```

```python
trial['birthdate'] = pd.to_datetime(trial['birthdate'])
```

```python
trial.info()
```

```python
trial_col =trial.columns
trial_col
```

```python
trial = trial.astype({
    'assigned_sex': 'category',
    'city': 'category',
    'state': 'category',
    'zip_code': 'category',
    'adverse_reaction': 'category',
    'dosage_type': 'category'
})
```

```python
trial.info()
```

```python
trial.patient_id.nunique()
```

```python
trial.assigned_sex.nunique()
```

```python
gender = trial.assigned_sex.value_counts().reset_index()
gender
```

```python
print(f'There are {gender.loc[0,'count']} Female and {gender.loc[1,'count']} Male.')
```

```python
plt.figure(figsize=(5,4))
ax = sns.countplot(trial, x='assigned_sex')
for bar in ax.containers:
    ax.bar_label(bar)
plt.title('Gender Frequency')
plt.xticks(['female', 'male'], ['Female', 'Male'])
plt.ylabel('Total participants')
plt.xlabel('Gender')
plt.show()
```

```python
#  to calculate age
# set refrence date : Jan 1, 2018
refrence_date=pd.Timestamp('2018-01-01')
refrence_date
```

```python
trial['age']=((refrence_date-trial['birthdate']).dt.days/365.25).astype('int')
```

```python
trial.groupby('dosage_type',observed=True)['adverse_reaction'].value_counts().unstack('dosage_type')
```
