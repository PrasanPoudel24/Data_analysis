# Frequency Counts
- it summarize how often each category or value appears in a datset foe a single variable
- it is foundation of EDA(Ezploratory Data Analysis) for categorical data
- it helps to find patterns , distributions or imbalance

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
titanic=sns.load_dataset('titanic')
titanic
```

# frequency count for passenger class

```python
titanic["class"].value_counts(normalize=True)*100
```

```python
sns.catplot(titanic,x='class',kind='count')
```

# Cross-Tabulation
- it extends frequency count by multiple categories

```python
pd.crosstab(titanic['class'],titanic['survived'],margins=True)
```

```python
pd.crosstab([titanic['class'],titanic['sex']],titanic['survived'],normalize='index')*100
```

```python
pd.crosstab([titanic['class'],titanic['sex']],titanic['survived'],normalize='columns')*100
```
