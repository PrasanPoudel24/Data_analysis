### What is Data Visualization in Data Analytics?
Data visualization is the graphical representation of data and information using visual elements like charts, graphs, maps, and dashboards. In data analytics, it's not just about making pretty pictures; it's about revealing patterns, trends, outliers, and relationships in data that might be hidden in spreadsheets or raw numbers.

- Core Purpose:
    - It helps analysts and stakeholders make sense of complex datasets quickly.
    - Humans process visuals faster than text.
    - Visualizations bridge the gap between data and decision-making.
- Key Benefits:

    - Insight Discovery: Spot correlations (e.g., sales increasing with marketing spend) or anomalies (e.g., a sudden drop in user engagement).
    - Storytelling: Communicate findings to non-technical audiences, like executives or clients.
    - Exploration: During analysis, visuals help iterate on hypotheses.
    - Efficiency: Reduces the time needed to understand data compared to poring over tables.

In analytics workflows, visualization fits into the "presentation" phase but also aids in exploratory data analysis (EDA) earlier on. For instance, in a retail analytics project, visualizing sales by region could reveal that one area outperforms others due to seasonal trends.
- Python libraries for creating these visuals in data analytics:

    - `Matplotlib`: A foundational plotting library offering fine-grained control over every aspect of a plot. It's versatile but requires more code for complex visuals.
    - `Seaborn`: Built on Matplotlib, it simplifies creating statistical and aesthetically pleasing visualizations with less code. It's ideal for quick, publication-quality plots.
    - `Pandas Plotting`: Built on Matplotlib, Pandas provides a convenient plotting interface directly from DataFrames, perfect for quick exploratory visuals.

```python
# Standard imports
import pandas as pd
import numpy as np

# Visualization library imports
import matplotlib.pyplot as plt
import seaborn as sns

# A special command to make plots appear inline in Jupyter Notebook
%matplotlib inline

# Set a default style for our plots for a nice look
sns.set_style('whitegrid')
```

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Product_A_Sales': [220, 250, 230, 280, 310, 290],
    'Product_B_Sales': [180, 200, 240, 210, 190, 220],
    'Customer_Satisfaction': [8.5, 8.2, 8.8, 9.0, 9.1, 8.9]
}

df = pd.DataFrame(data)

print("--- Our Sample Sales Dataset ---")
df
```

# line plot 
- this plot is used for univariate and bivariate analysis.
- this shows time series data over the period of time.
- suitable for showing market trend , product efficiency over time.
- `Question:`How did sales for product A and Product B change over the first six months?

```python
# figure for define size of picture
plt.figure(figsize=(8,4))

# plotting with matplotlib
plt.plot(df['Month'],df['Product_A_Sales'],marker='o',linestyle='--',label="Product A")
plt.title("Sales over first Six Months of Product A",fontsize=16,loc='right')
plt.xlabel("Months")
plt.ylabel("Sales in ($)")
plt.grid
# plt show
plt.show()
```

```python
# figure for define size of picture
plt.figure(figsize=(12,4))

# plotting with matplotlib
plt.plot(df['Month'],df['Product_A_Sales'],marker='o',linestyle='--',label="Product A")
plt.plot(df['Month'],df['Product_B_Sales'],marker='D',linestyle='-.',label="Product B")
plt.title("Monthly Sales Performance of Products",fontsize=16,loc='right')
plt.xlabel("Months")
plt.ylabel("Sales in ($)")
plt.legend()
#plt.grid
# plt show
plt.show()
```

# Box Plot
- univariate analysis and bivariate analysis with numeric or continuous data points.
- tp find outliers, variance, distribution
- `Question : ` distribution of total bill or Os there any outliers present in total_bill

```python
plt.figure(figsize=(15,5))
sns.boxplot(tips,x='total_bill',hue='size')
plt.show()
```

# Bar plot
- relationship between catgorical and numeric data points
- use for bi or multi-variate analysis
- it shows average value over categories
- `Question :` Total sales by Days

```python
plt.figure(figsize=(10,5))
new=sns.barplot(tips,x = 'day',y='total_bill',hue='smoker',errorbar='sd',capsize=0.1)
for bar in new.containers:
    new.bar_label(bar,fmt='%.2f',padding=-30,label_type='center')
plt.title('Total sales per day(non-smoker vs smoker)',fontsize=16)
plt.legend(title='smoker',loc='best',bbox_to_anchor=(1, 1, 0.15, 0))
plt.show()
```

# Pie chart
- A pie chart is a circular statistical graphic that divides a circle into slices to illustrate numerical proportions or percentages of a whole.
- to display the proportion of categories within a dataset(eg: per of tips by day)

```python
tip_by_day=tips.groupby('day',observed=True)['tip'].sum().reset_index()
tip_by_day
```

```python
# create a pie chart
plt.figure(figsize=(10,5))
plt.pie(tip_by_day['tip'],labels=tip_by_day['day'],
       autopct='%.2f%%',
       startangle=90,
       explode=[0,0.1,0,0],
        wedgeprops={
            'edgecolor':'black',
            'linewidth':1
        },
        colors=sns.color_palette('pastel')
       )
plt.legend()
plt.show()
```

# Scatter Plot
- A scatter plot is a type of data visualization that displays individual data points as dots on a two-dimensional plane.
- where x-axis and y axis represents two different variables.
- to show relationship between two numerical variables and optionally explore additional dimensions with color, size or shape.
- correlation, cluters, multi_variable data
- `Questions :` show correlation between total_bill and tip

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=tips,
               x='total_bill',
            y='tip',hue='sex',size='size',style='smoker',alpha=0.6
            )
plt.show()
```

```python
tips[tips['smoker']=='Yes'][['total_bill','tip']].corr()
```

```python
tips[tips['smoker']=='No'][['total_bill','tip']].corr()
```

# HeaTMap
- HeatMap is a data visualization technique that uses color intensity to represent the magnitude of values in a matrix or table
- often used for correlation matrics, confusion matrics or time_series data.

```python
corr_matrix=tips[['total_bill','tip','size']].corr()
corr_matrix
```

```python
# set the style
sns.set_style('white')

plt.figure(figsize=(10,5))
sns.heatmap(corr_matrix,annot=True,fmt='.2f',vmin=-1,vmax=1,center=0,square= True,
           
           cbar_kws = {'label':'correlation cofficient'})
plt.show()
```

# subplot
- to arrange multiple plots in agrid or table for comparision or to present related data
 

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10,10))

# subplot1 : scatter plot
sns.scatterplot(tips, x= 'total_bill', y='tip',hue= 'smoker', ax= axes[0])
axes[0].set_title('Scatter: total bill vs tip')
axes[0].set_xlabel('Total Bill($)')
axes[0].set_ylabel('Tip ($)')

# subplot2: Barplot
sns.barplot(tips, x= 'day', y='tip', hue='sex', ax= axes[1])
axes[1].set_title('Bar: Avg. tip by day')
axes[1].set_xlabel('Day')
axes[1].set_ylabel('Avg Tip ($)')

# subplot3 : pie chart
plt.pie(tip_by_day['tip'], labels=tip_by_day['day'], autopct='%.1f%%')
axes[2].set_title('Pie: Proportion of total tips by day')


plt.tight_layout()
plt.show()
```

# FacetGrid
- it creates a grid of subplots based on one or more categorical variable, enabling comparision across subsets of data.
- comparing distributions or relationships across categories
- multi_dimensional exploration

```python
# create a FacteGrid
g = sns.FacetGrid(tips, row ='time',col='day',hue='sex',height=3)

# map a scatter plot to each facet

g.map(plt.scatter, 'total_bill','tip')
g.add_legend(title='sex')
plt.show()
```
